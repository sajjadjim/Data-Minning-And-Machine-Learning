```python
x = [3.2, 9.8, 8.6, 3.5, 6.5, 2.5, 4.6]
y = [10, 18, 4, 25, 26, 27, 35]


y_min = min(y)
y_max = max(y)
y_normalized = [(yi - y_min) / (y_max - y_min) for yi in y]


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.bar(x, y_normalized, color='blue')
plt.title('Bar Chart of x and Normalized y')
plt.xlabel('x values')
plt.ylabel('Normalized y values')


plt.subplot(1, 2, 2)
plt.scatter(x, y_normalized, color='red')
plt.title('Scatter Plot of x and Normalized y')
plt.xlabel('x values')
plt.ylabel('Normalized y values')


plt.tight_layout()
plt.show()
```
