```python
##Scatter Plot  [ No connection between one point to another point]
#When object more , we better use to scatter plot

x = df['sepal_length']
y = df['sepal_width']

plt.scatter(x,y, color ='b')   # c that mean color
plt.show()
```

```python
## Two scatter plot making

x=df['sepal_length']
y=df['sepal_width']

x1 =df['petal_length']
y1=df['petal_width']

plt.scatter(x,y,c='g', label ='Sepal_Length & Seapl_width')
plt.scatter(x1,y1,c='r',label ='Petal_Length & Peatl_width')
plt.legend()
plt.show()
```

```python
x=df['sepal_length']
y=df['sepal_width']

x1 =df['petal_length']
y1=df['petal_width']

#Title the plot is given use this
plt.title('Sepal vs Petal')
plt.xlabel('X Axis')
plt.ylabel('Y axis')


plt.scatter(x,y,c='g', label ='Sepal_Length & Seapl_width')
plt.scatter(x1,y1,c='r',label ='Petal_Length & Peatl_width')
plt.legend()
plt.show()
```

```python
## Bar Plot create

x = [0.25, 1.25, 2.25, 3.25, 4.25]
y=[50,40,70,80,20]

x1 = [0.75, 1.75, 2.75, 3.75, 4.75]
y1 = [80,20,20,50,60]

plt.title('Information')
plt.xlabel('Days')
plt.ylabel('Distance(km)')


plt.bar(x1,y1, color='green',label='Audi',width=.5)
plt.bar(x,y, color='pink',label='BWM',width=.5)
plt.legend()
plt.show()
```

```python
### Pi Chart

data=df["species"].value_counts()

name =data.index
count=data.values

plt.title('Scores')
plt.xlabel('Names')
plt.ylabel('Total Cunt')

plt.barh(name,count , color='yellow')
plt.show()
```

```python
slices = [4,2,1,3]
activities = ['sleeping','eating','working','playing']
cols = ['c','g','r','b']  #Pie chart color represnt with "cols"

plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=45,
        shadow= True,
        explode=(0,0.1,0,0),   #Gap from another to another potion here use that
        autopct='%1.1f%%')

plt.title('Pie Plot')

plt.show()
```

```python
slices = [4,2,1,3]
activities = ['sleeping','eating','working','playing']
cols = ['c','g','r','b']  #Pie chart color represnt with "cols"

plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=90,
        shadow=False,
        explode=(0,0.1,0,0),   #Gap from another to another potion here use that
        autopct='%1.0f%%',
        radius=1.2)

plt.title('Pie Plot')

plt.show()
```
