#Question
1 table healt data select age,avg-glucose,bmi and 25% quartile vale with std

2.take healt data set make visualization on those data by selecting subplot

```python
selected_data = df[['age', 'avg_glucose_level', 'bmi']]

quantile_25 = selected_data.quantile(0.25)
print("25th percentile:\n", quantile_25)

std_dev = selected_data.std()
print("\nStandard Deviation:\n", std_dev)
```

```python
import matplotlib.pyplot as plt

selected_data = df[['age', 'avg_glucose_level', 'bmi']]

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(selected_data['age'], bins=20, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

plt.subplot(2, 2, 2)
plt.scatter(selected_data['age'], selected_data['avg_glucose_level'], color='green')
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.title('Age vs. Average Glucose Level')

plt.subplot(2, 2, 3)
plt.boxplot(selected_data['bmi'], vert=False)
plt.xlabel('BMI')
plt.title('BMI Distribution')

plt.subplot(2, 2, 4)
plt.hist(selected_data['avg_glucose_level'], bins=20, color='orange')
plt.xlabel('Average Glucose Level')
plt.ylabel('Frequency')
plt.title('Average Glucose Level Distribution')

plt.tight_layout()
plt.show()

```
